```python
import numpy as np

# Input
x1,x2,x3,x4 = 1,1,0,1
Y = 1

epochs = 4
alpha = 2.0      # Strong learning rate for fast convergence


# Initial Weights (Balanced)

w15,w25,w35,w45 = 0.5,-0.4,0.2,0.3
w16,w26,w36,w46 = 0.4,0.5,-0.3,0.6

w57,w67 = -0.6,0.7


# Bias

t5,t6,t7 = 0.2,0.2,-0.4


# Sigmoid

def sigmoid(x):
    return 1/(1+np.exp(-x))


for epoch in range(epochs):

    print("\nEpoch",epoch+1)

    # -------- Forward Pass --------

    O5 = sigmoid(x1*w15+x2*w25+x3*w35+x4*w45+t5)

    O6 = sigmoid(x1*w16+x2*w26+x3*w36+x4*w46+t6)

    O7 = sigmoid(O5*w57 + O6*w67 + t7)


    # -------- Error --------

    error = Y - O7

    print("Output =",round(O7,4),
          " Error =",round(abs(error),4))


    # -------- Backpropagation --------

    E7 = O7*(1-O7)*error

    E6 = O6*(1-O6)*(E7*w67)

    E5 = O5*(1-O5)*(E7*w57)


    # -------- Update Hidden → Output --------

    w57 += alpha*E7*O5
    w67 += alpha*E7*O6


    # -------- Update Input → Hidden --------

    w15 += alpha*E5*x1
    w25 += alpha*E5*x2
    w35 += alpha*E5*x3
    w45 += alpha*E5*x4

    w16 += alpha*E6*x1
    w26 += alpha*E6*x2
    w36 += alpha*E6*x3
    w46 += alpha*E6*x4


    # -------- Update Bias --------

    t5 += alpha*E5
    t6 += alpha*E6
    t7 += alpha*E7


print("\nTraining Completed")

print("Final Output =",round(O7,4))
print("Final Error =",round(abs(error),4))
     
```
