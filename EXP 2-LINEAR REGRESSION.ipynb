```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Midterm (X) and Final (y)
X = np.array([72,50,81,74,94,86,59,83,65,33,88,81]).reshape(-1,1)
y = np.array([84,63,77,78,90,75,49,79,77,52,74,90])

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Get slope and intercept
b = model.coef_[0]
a = model.intercept_

print("Slope (b):", b)
print("Intercept (a):", a)

# Predict for 86
prediction = model.predict([[86]])
print("Predicted Final Grade:", prediction[0])

# ---- Plotting ----
plt.scatter(X, y, color='blue', label='Actual Data')

# Regression line
x_line = np.linspace(min(X), max(X), 100).reshape(-1,1)
y_line = model.predict(x_line)
plt.plot(x_line, y_line, color='red', label='Regression Line')

# Predicted point
plt.scatter(86, prediction, color='green', s=100, label='Prediction (86)')

plt.xlabel("Midterm Marks")
plt.ylabel("Final Exam Marks")
plt.title("Least Squares Linear Regression")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Dataset
X = np.array([29, 15, 33, 28, 39]).reshape(-1, 1)
y = np.array([0, 0, 1, 1, 1])   # 0 = Fail, 1 = Pass

# Create and train model
model = LogisticRegression()
model.fit(X, y)

# Get learned parameters
intercept = model.intercept_[0]
coef = model.coef_[0][0]

print("Intercept:", intercept)
print("Coefficient:", coef)

# 1️⃣ Probability for 33 hours
prob_33 = model.predict_proba([[33]])[0][1]
print("Probability of passing for 33 hours:", prob_33)

# 2️⃣ Minimum hours for probability > 95%
target_prob = 0.95
z_required = np.log(target_prob / (1 - target_prob))

# Solve z = intercept + coef * hours
hours_required = (z_required - intercept) / coef
print("Minimum hours for >95% probability:", np.ceil(hours_required))

# ---- Plot sigmoid curve ----
x_values = np.linspace(10, 45, 200).reshape(-1, 1)
y_values = model.predict_proba(x_values)[:, 1]

plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(x_values, y_values, color='red', label='Logistic Curve')

plt.xlabel("Hours Studied")
plt.ylabel("Probability of Pass")
plt.title("Logistic Regression")
plt.legend()
plt.grid(True)
plt.show()
```
